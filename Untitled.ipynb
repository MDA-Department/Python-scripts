```python
import pandas as pd

tcount = 0
tname_in_emails = 0
tpreffered = 0

def run_file(filename):
    df = pd.read_csv(filename)
    df['splits'] = df['email'].str.split('@').str[0]
    df=df.applymap(str)
    df.dtypes
    print(f"Loaded file: {filename}")
    
    count = 0
    name_in_emails = 0
    preffered = 0

    for index, row in df.iterrows():
        count+=1
        fname = str(row['firstname']).strip().replace('-', '').replace('.', '')[:6]
        lname = str(row['lastname']).strip().replace('-', '').replace('.', '')[:6]
        
        if row['ispreferred'] == 't':
            preffered+=1
            if fname in row['splits'] or lname in row['splits']:
                name_in_emails+=1


    print("Run:")
    print(count)
    print(preffered)
    print(name_in_emails)
    
    global tcount
    global tname_in_emails
    global tpreffered
    
    tcount+=count
    tname_in_emails+=name_in_emails
    tpreffered+=preffered

run_file('data-1.csv')
run_file('data-2.csv')
run_file('data-3.csv')
run_file('data-4.csv')
run_file('data-5.csv')

print("TOTALS:")
print(tcount)
print(tpreffered)
print(tname_in_emails)
    
```
