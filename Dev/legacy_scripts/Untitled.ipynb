```python

# Python program to read
# json file
 
import pandas as pd
import xlsxwriter
import json
import requests



url = requests.get("http://sis.rutgers.edu/soc/api/courses.json?year=2022&term=1&campus=NB")
text = url.text

data = json.loads(text)




# Opening JSON file
#f = open('NB.json')
 
# returns JSON object as
# a dictionary
#data = json.load(f)
 
# Iterating through the json
# list
out = []
data_count = 0
match_count = 0

for i in data:
    data_count = data_count + 1
    #Vars reset
    coursestring_id = ""
    location_out = ""
    insttext_out = ""
    meeting_out = ""
    starttime_out = ""
    endtime_out = ""
    campusname_out = ""
    buildcode_out = ""
    roomnum_out = ""
    coursenum_out = ""
    subject_out = ""
    title_out = ""
    meetmode_out = ""
    
    # ------- ID FIELDS -------------
    # COURSE ID
    coursenum_out = i["courseNumber"]
    subject_out = i["subjectDescription"]
    title_out = i["expandedTitle"]
    
    # CAMPUS LOCATION
    location_list = i["campusLocations"]
    icount = 0
    for item in location_list:
        location = item["description"]
        if location in ["Busch", "C/D", "College Avenue", "Downtown New Brunswick", "Livingston"]:
            if icount == 0:
                location_out = location_out + location
                icount = 1
            else:
                location_out = location_out + "/" + location

    # -------- SECTIONS -------           
    section_list = i["sections"]
    for item in section_list:
        # SECTION ID
        section_id = item["index"]
        
        # INSTRUCTORS TEXT
        insttext_out = item["instructorsText"]
        
        
        # MEETING TIME
        meetingTimes = item["meetingTimes"]
        for data in meetingTimes:
            if "meetingDay" in data:
                meeting_out = data["meetingDay"]
            
            if "startTimeMilitary" in data:
                starttime_out = data["startTimeMilitary"]
            
            if "endTimeMilitary" in data:
                endtime_out = data["endTimeMilitary"]
            
            if "campusName" in data:
                campusname_out = data["campusName"]
            
            if "buildingCode" in data:
                buildcode_out = data["buildingCode"]
            
            if "roomNumber" in data:
                roomnum_out = data["roomNumber"]
            
            if "meetingModeDesc" in data:
                meetmode_out = data["meetingModeDesc"]

            
            
            if location_out:
                match_count = match_count + 1
                out.append({"courseNumber":coursenum_out,
                            "subjectDescription":subject_out,
                            "expandedTitle":title_out,
                            "campusLocation":location_out,
                            "section_id":section_id,
                            "instructorsText":insttext_out,
                            "meetingDay":meeting_out,
                            "startTimeMilitary":starttime_out,
                            "endTimeMilitary":endtime_out,
                            "campusName":campusname_out,
                            "buildingCode":buildcode_out,
                            "roomNumber":roomnum_out,
                            "meetingModeDesc":meetmode_out})
            


out_df = pd.DataFrame(out)
            
#output file variables
filename = ('outputjson.xlsx')
writer = pd.ExcelWriter(filename)


#Write dataframe to excel minus headers
out_df.to_excel(writer, engine='xlsxwriter', index=False)
writer.save()

# Closing t file
#f.close()

# Validation Statements
print("Number of data points in json: {}".format(data_count))
print("Number of matches: {}".format(match_count))
print(out)

```
