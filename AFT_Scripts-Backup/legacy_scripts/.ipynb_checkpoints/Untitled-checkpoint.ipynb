```python

# Python program to read
# json file
 
 
import json
 
# Opening JSON file
f = open('NB.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
 
# Iterating through the json
# list
out = []
data_count = 0
match_count = 0

for i in data:
    data_count = data_count + 1
    #Vars reset
    coursestring_id = ""
    location_out = ""
    insttext_out = ""
    meeting_out = ""
    starttime_out = ""
    endtime_out = ""
    campusname_out = ""
    buildcode_out = ""
    roomnum_out = ""
    coursenum_out = ""
    subject_out = ""
    title_out = ""
    meetmode_out = ""
    
    coursestring_id = i["courseString"]
    
    # CAMPUS LOCATION
    location_list = i["campusLocations"]
    icount = 0
    for item in location_list:
        location = item["description"]
        if location in ["Busch", "C/D", "College Avenue", "Downtown New Brunswick", "Livingston"]:
            if icount == 0:
                location_out = location_out + location
                icount = 1
            else:
                location_out = location_out + "/" + location

    # INSTRUCTORS TEXT
    section_list = i["sections"]
    print(len(section_list))
    for item in section_list:
        section = item["instructorsText"]
        insttext_out = insttext_out + section + " / "

        

    
    
    
    
    if location_out:
        match_count = match_count + 1
        out.append({"coursestring_id":coursestring_id,
                    "campusLocation":location_out,
                    "instructorsText":insttext_out,
                    "meetingDay":meeting_out,
                    "startTimeMilitary":starttime_out,
                    "endTimeMilitary":endtime_out,
                    "campusName":campusname_out,
                    "buildingCode":buildcode_out,
                    "roomNumber":roomnum_out,
                    "courseNumber":coursenum_out,
                    "subjectDescription":subject_out,
                    "expandedTitle":title_out,
                    "meetingModeDesc":meetmode_out})
            

# for d in out:
#     location = out["campusLocations"]

print("Number of data points in json: {}".format(data_count))
print("Number of matches: {}".format(match_count))
print(out)

 
# Closing file
f.close()
```
